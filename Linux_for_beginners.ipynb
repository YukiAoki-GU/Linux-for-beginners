そのためMac OSを使っている人はデフォルトでLinuxが使えると思ってもらって構いません。

## シェルとは？
Linuxのシステムを操作するための言語だと思ってください。
Linuxのスタンダードは[bash](https://www.tohoho-web.com/ex/shell.html)というシェルです。
何のシェルを使っているか調べるには


```python
echo $SHELL
```

シェルを使ってif文，for文，while文などの命令を書くことができます。
以下はbashでの0-9を標準出力に表示する例です。

```python
#!/usr/bin/bash
for ((i=0; i < 10; i++)); do
    echo $i
done
```

## ディレクトリの構成
ディレクトリとはWindowsでいうフォルダと同じ意味です。
Windowsではフォルダの区切り文字には'¥'を使いますが，Linuxでは'/'を使います。
ファイル操作の基本的なコマンドは[こちら](http://ss.sguc.ac.jp/~rider/unix/linuxfile.html)，ディレクトリ操作の基本的なコマンドは[こちら](http://ss.sguc.ac.jp/~rider/unix/linuxdirectory.html)をみてください。
Linuxのディレクトリの構造は大まかに以下のようになっています。
- /bin ：binaryの略。標準コマンドが置いてある。(例：cd, cp, mv)
- /sbin：system binaryの略。システム管理用(ルートが利用する)コマンドが置いてある
- /boot：システムの起動に必要なものが置いてある
- /etc ：エトセトラの略。設定ファイル(*.confとか)とかプログラム本体が置いてある
- /home：ユーザーのディレクトリがある
- /lib ：ライブラリの略。共有ライブラリが置いてある
- /dev ：デバイスの略
- /opt ：オプショナルの略。ユーザーが自分でソフトをインストールするときに使う場所かな・・・
- /var ：variableの略。頻繁に更新されるログとかが置いてある
- /proc：プロセスの略。メモリなどのリソースのパラメータが置いてある
- /usr ：ユーザーの略。その下にbin，lib，sbinなどがある。

### パスについて
Linuxのコマンドがどのディレクトリに置いてあるかを知りたい場合はwhichを使います。

例：echoコマンドはどこにあるの？

```python
which echo
```

パスがどこに張られているかを知りたい場合は

```python
echo $PATH
```

パスを追加したい場合は~/.bashrcに以下を追加

```python
export PATH=$PATH:追加するディレクトリ
```

その後，.bashrcの設定を反映させる

```python
source ~/.bashrc
```
